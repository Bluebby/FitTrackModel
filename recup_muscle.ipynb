```python
import os
import re

folder_path = "data/body"  # À adapter selon ton dossier

# Expression régulière pour capturer uniquement le nom anatomique
pattern = re.compile(r"^FJ\d+_BP\d+_FMA\d+_(.*)\.obj$", re.IGNORECASE)

cleaned_names = []

for filename in os.listdir(folder_path):
    if filename.endswith(".obj"):
        match = pattern.match(filename)
        if match:
            cleaned_names.append(match.group(1))

# Affichage des noms nettoyés
for name in cleaned_names:
    print(name)
```
